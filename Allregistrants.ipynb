```python
# import seaborn library
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Try exception error
try:
    df = pd.read_csv("AllRegistrantsToUpendra_20190219.csv", encoding = "ISO-8859-1")
except UnicodeDecodeError as error:
    print("UnicodeDEcodeErorr")
except FileNotFoundError as fnf_error:
    print("File not found")
```

```python
# Drop the columns that are not useful
df.drop(['Timestamp', 'Waiver 1', 'Waiver 2', 'SWD?'], axis = 1, inplace=True)
df.shape
```

```python
# Make a list of old column names
old_columns = list(df.columns)
```

```python
# Make a list of new column names
new_columns = ['Accepted', 'Confirmed', 'Occupation', 'Organization', 'Department', 'Research', 'Reason', 'OS', 'Manage_data', \
           'Analyze_data', 'Frequency', 'Complexity', 'Bash', 'Directory_task', 'Version_control', 'Version_control_task',\
          'Python_Jupyter', 'Program_task', 'Future_workshop']
```

```python
# Rename the columns
col_rename_dict = {i:j for i, j in zip(old_columns, new_columns)}
df_new = df.rename(columns=col_rename_dict)
```

### Research

```python
# Import libraries
from wordcloud import WordCloud,STOPWORDS
from io import StringIO
```

```python
si = StringIO()
df_new["Research"].apply(lambda x: si.write(str(x)))
s = si.getvalue()
si.close()
```

```python
#set the stopwords list
stopwords= set(STOPWORDS)
new_words = ['currently', 'research', 'study', 'testing', 'lab', 'interested', 'using', 'work', 'use']
final_words = stopwords.union(new_words)
```

```python
# Read the whole text.
text = s

# Generate a word cloud image
wordcloud = WordCloud(background_color="white",
                     max_words=2000, 
                     min_font_size =12, 
                     max_font_size=20, 
                     relative_scaling = 0.5, 
                     stopwords=STOPWORDS,
                     normalize_plurals= True).generate(text)

# take relative word frequencies into account, lower max_font_size
wordcloud = WordCloud(background_color="white",max_words=len(s),max_font_size=40, relative_scaling=.5).generate(text)
plt.figure();
plt.imshow(wordcloud);
plt.axis("off");
```
